```python
import matplotlib.pyplot as plt
import pandas as pd

PATH_OUT = input('Введите PATH_OUT')

df = pd.read_csv(PATH_OUT + '/lr.csv')

def least_squares(x, y):
    n = len(x)
    avex = sum(x)/n
    avey = sum(y)/n
    numerator = sum([(x[i] - avex) * (y[i] - avey) for i in range(n)])
    denominator = sum([(x[i] - avex)**2  for i in range(n)])
    k = numerator / denominator
    b = avey - k * avex
    return (k, b)
     

slope, intercept = least_squares(df['GRE Score'], df['Chance of Admit'])
line = slope * df['GRE Score'] + intercept
plt.scatter(df['GRE Score'], df['Chance of Admit'])
plt.plot(df['GRE Score'], line, color='red', label='Line of Best Fit')
plt.show()
```
